```python
import pandas as pd
```

```python
train_df = pd.read_csv("/content/drive/MyDrive/Data_sml_kaggle1/Train.csv")
train_df
```

```python
test_df = pd.read_csv("/content/drive/MyDrive/Data_sml_kaggle1/Test.csv")
test_df
```

```python
train_df.columns
```

```python
X_train = train_df.drop(['Id', 'target'], axis=1)
y_train = train_df['target']

print(X_train.shape)
print(y_train.shape)
```

```python
X_test = test_df.drop(['Id'], axis=1)

print(X_test.shape)
X_test
```

```python
print(X_train.shape)
X_train
```

```python
#Standardization of the columns on the basis of Min-Max Scaling 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Normalizing train data
X_train_new = pd.DataFrame(scaler.fit_transform(X_train))
X_train_new.columns = X_train.columns
X_train = X_train_new
```

```python
# Normalizing test data
X_test_new = pd.DataFrame(scaler.fit_transform(X_test))
X_test_new.columns = X_test.columns
X_test = X_test_new
# X_test
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
```

### Recursive feature elimination

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
rfe = RFE(logreg)
sf = rfe.fit(X_train, y_train)
sc = X_train.columns[rfe.get_support(indices=True)]
X_train = X_train[sc]
X_train
```

```python
X_test.columns
```

```python
lst = []
for col in X_train.columns:
  lst.append(col)
print(lst)

drop_cols = []
for col in X_test.columns:
  # print(col)
  if(col in lst):
    pass
  else:
    drop_cols.append(col)

print(drop_cols)


X_test = X_test.drop(drop_cols, axis=1)

print(X_test.shape)
X_test
```

## Model : Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
logreg = LogisticRegression(solver='newton-cg')
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)

# y_pred_valid = logreg.predict(X_valid)
# print(accuracy_score(y_valid, y_pred_valid)*100)
```

```python
y_pred
```

### prediction file generation for Logistic Regression

```python
y_pred_df = pd.DataFrame(y_pred, columns=['target']) 

id_vals = [num for num in range(len(y_pred))]
# print(len(id_vals), id_vals)

y_pred_df.insert(0, "ID", id_vals)
# y_pred_df

y_pred_df.to_csv("pred_lr.csv", encoding='utf-8', index=False)
```

####Saves the logistic Regression model

```python
import joblib
```

```python
# Save the model as a .pkl file using pickle library
joblib.dump(logreg, "logistic_reg.pkl")

# Loading the model from the logistic_reg.pkl
logreg_loaded = joblib.load("logistic_reg.pkl")
y_pred2 = logreg_loaded.predict(X_test)
print(y_pred==y_pred2)
```
